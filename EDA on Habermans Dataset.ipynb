# Haberman Cancer Survival Dataset

   Haberman Cancer Survival Dataset contains data of patients who had the surgery of breast cancer, This was the study conducted by **University of Chicago’s Billings Hospital** between year 1958 to 1970. The study was done using 4 attribute on 305 individuals based on the following factors:
    1. Age (30 to 83)
    2  Operation Year(1958–1969)
    3. Number of Axillary nodes[Lymph Nodes] (0–52)
    4. Survival Status (checks if patients survived 5 years)

**CDF:**
     
   The CDF for the Petal Length for different ***bin*** values are shown below:
   - The Survival_Status can be measure using the alive or dead state. 
   Let us consider live to be 0 and dead to be 1.

```python
live=haberman.loc[haberman["Survival_Status"]==1]
dead=haberman.loc[haberman["Survival_Status"]==2]
```

```python
counts, bin_edges = np.histogram(live['Lymph_Nodes'], bins=20,density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.legend(['1 ','2 '])
plt.title('CDF of people alive')
plt.xlabel('Lymph_Nodes_live')
plt.show();

print("PDF:",pdf);
print("bin_edges",bin_edges)
```

```python
counts, bin_edges = np.histogram(dead['Lymph_Nodes'], bins=20,density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.legend(['1 ','2 '])
plt.title('CDF of people dead')
plt.xlabel('Lymph_Nodes_dead')
plt.show();

print("PDF:",pdf);
print("bin_edges",bin_edges)
```

**Observations:** Above is the plot for CDF where 1 is the PDF and 2 is bin_edge values

**Median, Percentile, Quantile, IQR, MAD:**

```python
print("\nMedians:")
print(np.median(live['Lymph_Nodes']))
print(np.median(dead['Lymph_Nodes']))

print("\nQuantiles:")
print(np.percentile(live['Lymph_Nodes'],np.arange(0, 100, 25)))
print(np.percentile(dead['Lymph_Nodes'],np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print(np.percentile(live['Lymph_Nodes'],90))
print(np.percentile(dead['Lymph_Nodes'],90))

from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(live['Lymph_Nodes']))
print(robust.mad(dead['Lymph_Nodes']))
```
